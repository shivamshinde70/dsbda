```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df=pd.read_csv("E:\heartdisease.csv")
```

```python
df
```

```python
df.dropna()
```

```python
#scatter
x=df['Age']
y=df['Slope']
plt.xlabel('Age')
plt.ylabel('Slope')
plt.title('Age Vs Slope')
plt.scatter(x,y)
plt.grid()
plt.show()
```

```python
#pie
x=df['Age']
plt.xlabel('Age')
plt.title('Age')
plt.pie(x)
plt.grid()
plt.show()
```

```python
#line
x=df['Age']
y=df['Slope']
plt.xlabel('Age')
plt.ylabel('Slope')
plt.title('Age Vs Slope')
plt.plot(x,y,marker='o',color='b')
plt.grid()
plt.show()
```

```python
#Histogram
x=df['Age']
plt.xlabel('Age')
plt.title('Age')
plt.hist(x)
plt.show()
```

```python
#bar
x=df['Age']
y=df['Slope']
plt.xlabel('Age')
plt.ylabel('Slope')
plt.title('Age Histogram')
plt.bar(x,y,color=['r','b'])
plt.show()
```

```python
#Box
x=df['Age']
plt.xlabel('Age')
plt.title('Age Piechart')
plt.boxplot(x)
plt.show()
```
